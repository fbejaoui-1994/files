```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Read data from CSV file
data = pd.read_excel('/content/StudentsPerformance.xlsx')  # Replace 'your_file.csv' with your actual file path

# Step 2: Choose the column to analyze (replace 'your_column' with the actual column name)
column_name = 'reading score*'# Replace this with the actual name of your column
column_data = data['writing score']

# Step 3: Calculate the mean and standard deviation
mean = np.mean(column_data)
std_dev = np.std(column_data)

# Step 4: Calculate the percentage of data within 1, 2, and 3 standard deviations
within_1_std = np.sum((column_data >= mean - std_dev) & (column_data <= mean + std_dev)) / len(column_data) * 100
within_2_std = np.sum((column_data >= mean - 2*std_dev) & (column_data <= mean + 2*std_dev)) / len(column_data) * 100
within_3_std = np.sum((column_data >= mean - 3*std_dev) & (column_data <= mean + 3*std_dev)) / len(column_data) * 100

# Step 5: Print results
print(f"Percentage within 1 standard deviation: {within_1_std:.2f}%")
print(f"Percentage within 2 standard deviations: {within_2_std:.2f}%")
print(f"Percentage within 3 standard deviations: {within_3_std:.2f}%")

# Visualizing the distribution and standard deviation intervals
plt.hist(column_data, bins=30, alpha=0.7, color='blue', edgecolor='black')
plt.axvline(mean, color='red', linestyle='dashed', linewidth=1, label='Mean')
plt.axvline(mean - std_dev, color='green', linestyle='dashed', linewidth=1, label='Mean - 1 Std Dev')
plt.axvline(mean + std_dev, color='green', linestyle='dashed', linewidth=1, label='Mean + 1 Std Dev')
plt.axvline(mean - 2*std_dev, color='orange', linestyle='dashed', linewidth=1, label='Mean - 2 Std Dev')
plt.axvline(mean + 2*std_dev, color='orange', linestyle='dashed', linewidth=1, label='Mean + 2 Std Dev')
plt.axvline(mean - 3*std_dev, color='purple', linestyle='dashed', linewidth=1, label='Mean - 3 Std Dev')
plt.axvline(mean + 3*std_dev, color='purple', linestyle='dashed', linewidth=1, label='Mean + 3 Std Dev')

plt.legend()
plt.title(f"Histogram of {column_name} with Standard Deviation Intervals")
plt.xlabel(column_name)
plt.ylabel("Frequency")
plt.show()

# Step 6: Check if the Empirical Rule is followed
empirical_rule_check = (
    abs(within_1_std - 68) <= 5 and
    abs(within_2_std - 95) <= 5 and
    abs(within_3_std - 99.7) <= 5
)

if empirical_rule_check:
    print("The data follows the Empirical Rule!")
else:
    print("The data does not follow the Empirical Rule.")
```
