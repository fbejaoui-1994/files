```python
q1=np.percentile(data['height'], 25)
q2=np.percentile(data['height'], 50)
q3=np.percentile(data['height'], 75)
```

```python
iqr=q3-q1
```

```python
print(f"Q1: {q1}")
print(f"Q2: {q2}")
print(f"Q3: {q3}")
print(f"IQR: {iqr}")
```

```python
upper_bound = q3 + 1.5 * iqr
lower_bound = q1 - 1.5 * iqr
```

```python
mean_height = data['height'].mean()
median_height = data['height'].median()
std_height = data['height'].std()
```

```python
outliers = data[(data['height'] > upper_bound) | (data['height'] < lower_bound)]
```

```python
print(outliers)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
# Calculate key statistics
minimum = data['height'].min()
maximum = data['height'].max()
# Calculate Pearson's Coefficient of Skewness
skewness = (3 * (mean_height - median_height)) / std_height
# Determine skewness type
if skewness > 0:
    skew_text = "Right-Skewed (Positive Skewness)"
elif skewness < 0:
    skew_text = "Left-Skewed (Negative Skewness)"
else:
    skew_text = "Symmetric (No Skewness)"

# Set a modern theme
sns.set_theme(style="whitegrid", palette="pastel")

# Create the boxplot with enhanced aesthetics
plt.figure(figsize=(10, 6))
boxplot = sns.boxplot(x=data['height'], color="lightblue", width=0.5, fliersize=5, linewidth=2.5,
                      boxprops=dict(facecolor='lightblue', edgecolor='darkblue', linewidth=2.5),
                      whiskerprops=dict(color='darkblue', linewidth=2.5),
                      capprops=dict(color='darkblue', linewidth=2.5),
                      medianprops=dict(color='red', linewidth=2.5))

# Add text annotations with improved positioning and formatting
plt.text(q1, -0.2, f"Q1: {q1:.0f}", ha='center', fontsize=12, color='darkblue', bbox=dict(facecolor='white', alpha=0.8, edgecolor='none'))
plt.text(median_height, -0.25, f"Median: {median_height:.0f}", ha='center', fontsize=12, color='red', bbox=dict(facecolor='white', alpha=0.8, edgecolor='none'))
plt.text(q3, -0.2, f"Q3: {q3:.0f}", ha='center', fontsize=12, color='darkblue', bbox=dict(facecolor='white', alpha=0.8, edgecolor='none'))
plt.text(minimum, -0.3, f"Min: {minimum:.0f}", ha='center', fontsize=12, color='green', bbox=dict(facecolor='white', alpha=0.8, edgecolor='none'))
plt.text(maximum, -0.3, f"Max: {maximum:.0f}", ha='center', fontsize=12, color='green', bbox=dict(facecolor='white', alpha=0.8, edgecolor='none'))

# Display skewness result above the boxplot
plt.text(q3 + iqr, 0.4, f"Skewness: {skewness:.2f}\n{skew_text}", ha='center', fontsize=14, color='purple',
         bbox=dict(facecolor='white', alpha=0.8, edgecolor='none'))

# Format x-axis tick labels to avoid scientific notation
plt.ticklabel_format(style='plain', axis='x')

# Set plot title and labels with improved formatting
plt.title('Enhanced Boxplot of Height (with Pearson Skewness)', fontsize=16, fontweight='bold', pad=20)
plt.xlabel('Height', fontsize=14, labelpad=10)
plt.ylabel('')

# Adjust y-axis limits to make space for annotations
plt.ylim(-0.35, 0.45)

# Remove y-axis ticks and labels for a cleaner look
plt.yticks([])

# Add a light grid for better readability
plt.grid(axis='x', linestyle='--', alpha=0.6)

# Show the plot
plt.show()
```

```python
# Create the KDE plot with enhancements
plt.figure(figsize=(10, 6))
sns.kdeplot(data['height'], color='blue', fill=True, alpha=0.6, label='height Distribution')  # Fill under the curve

# Add vertical lines for key statistics
plt.axvline(mean_height, color='red', linestyle='--', label=f'Mean: {mean_height:.0f}')
plt.axvline(median_height, color='green', linestyle='--', label=f'Median: {median_height:.0f}')
plt.axvline(q1, color='purple', linestyle=':', label=f'Q1: {q1:.0f}')
plt.axvline(q3, color='orange', linestyle=':', label=f'Q3: {q3:.0f}')

# Add annotations for key statistics
plt.text(mean_height, 0.0001, f'Mean: {mean_height:.0f}', rotation=90, verticalalignment='bottom', color='red', fontsize=10)
plt.text(median_height, 0.0001, f'Median: {median_height:.0f}', rotation=90, verticalalignment='bottom', color='green', fontsize=10)
plt.text(q1, 0.0001, f'Q1: {q1:.0f}', rotation=90, verticalalignment='bottom', color='purple', fontsize=10)
plt.text(q3, 0.0001, f'Q3: {q3:.0f}', rotation=90, verticalalignment='bottom', color='orange', fontsize=10)

# Add title and labels
plt.title('Distribution of height (KDE)', fontsize=16, fontweight='bold')
plt.xlabel('height', fontsize=14)
plt.ylabel('Density', fontsize=14)

# Format x-axis tick labels to avoid scientific notation
plt.ticklabel_format(style='plain', axis='x')

# Add grid for better readability
plt.grid(True, linestyle='--', alpha=0.6)

# Add legend
plt.legend(loc='upper right', fontsize=12)

# Show the plot
plt.show()
```
